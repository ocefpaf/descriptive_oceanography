```python
import gsw
import numpy as np
import matplotlib.pyplot as plt
```

```python
Te = np.arange(0, 32, 2)
Se = np.arange(32, 38.25, 0.25)

Sg, Tg = np.meshgrid(Se, Te)
levels = np.arange(20, 34, 0.5)

sigma_theta = gsw.sigma0_pt0_exact(Sg, Tg)

fig, ax = plt.subplots(figsize=(10, 10))
ax.set_ylabel(u"Temperatura \u00b0C")
ax.set_xlabel(r"Salinidade [g kg$^{-1}$]")

cs = ax.contour(Se, Te, sigma_theta, colors='black', levels=levels)
ax.clabel(cs, fontsize=9, inline=1, fmt='%2.1f')
ax.axis([31.8, 37.2, 0.0, 30.0])
fig.savefig("Diagrama-TS.pdf", dpi=150)
```
